```python
train_data = pd.read_csv("/kaggle/input/titanic/train.csv")
train_data.head()
```

```python
test_data = pd.read_csv("/kaggle/input/titanic/test.csv")
test_data.head()
```

```python
print(train_data.groupby("Sex")["Survived"].mean())
```

```python
test_predictions = test_data["Sex"].apply(lambda x: 1 if x == "female" else 0)

output = pd.DataFrame({
    "PassengerId": test_data["PassengerId"],
    "Survived": test_predictions
})
output.to_csv("baseline_submission.csv", index=False)
print("Baseline rule-based submission saved as 'baseline_submission.csv'")
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, f1_score

train_pred = train_data["Sex"].apply(lambda x: 1 if x == "female" else 0)

accuracy = accuracy_score(train_data["Survived"], train_pred)
f1 = f1_score(train_data["Survived"], train_pred)
conf_matrix = confusion_matrix(train_data["Survived"], train_pred)

print("=== Baseline Rule Evaluation on Training Set ===")
print(f"Accuracy: {accuracy:.4f}")
print(f"F1 Score: {f1:.4f}")
print("Confusion Matrix:")
print(conf_matrix)
```
